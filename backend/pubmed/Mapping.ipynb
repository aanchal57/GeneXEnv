```python
import pandas as pd
import csv
```

```python
df=pd.read_csv('/content/geneextractiondata-snpedia - cleaned-data.csv')
```

```python
len(df['genes'])
file = open('Gene_env.csv', 'w', newline ='')
 
with file:
    # identifying header 
    header = ['Gene', 'Env']
    writer = csv.DictWriter(file, fieldnames = header)
     
    # writing data row-wise into the csv file
    writer.writeheader()
    
```

```python
lis=['SNP','variant','interaction','diet','physical activity','excercise','alcohol','tobacco','cigarette']
filename = open('Gene_env.csv', 'a+', newline ='')
write = csv.writer(filename)
for i in range(0,len(lis)):
  for j in range(0,len(df['genes'])):
      write.writerow([df['genes'][j],lis[i]])
    
```
